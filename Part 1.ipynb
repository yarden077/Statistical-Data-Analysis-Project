# Chosen data set is: Cardiovascular Diseases Risk Prediction

URL: https://www.kaggle.com/datasets/alphiree/cardiovascular-diseases-risk-prediction-dataset

#### Data is about U.S. residents answering health-related telephone surveys that collect their health-related risk behaviors, chronic health conditions, and use of preventive services.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the csv file
df = pd.read_csv('CVD_cleaned.csv')
```

```python
# Choosing relevant columns:
df = df[['Depression', 'Sex', 'Weight_(kg)', 'Alcohol_Consumption', 
         'Green_Vegetables_Consumption', 'FriedPotato_Consumption' ]]

# Changing categorical variables into binary values
df['Depression'] = df['Depression'].replace(['Yes', 'No'], [1, 0])
df['Sex'] = df['Sex'].replace(['Male', 'Female'], [1, 0])
```

### Columns Explanation:

* Depression: Binary values (1 stands for Yes).
* Sex: Binary values (1 stands for Male).
* Weight: Continuous value, measured in kilograms.
* Alcohol Consumption: Continuous value, unit of measurement is the number of consumption in a month.
* Green Vegetables Consumption: Continuous value, unit of measurement is the number of consumption in a month.
* Fried Potato Consumption: Continuous value, unit of measurement is the number of consumption in a month.


 * #####  Alcohol Consumption, Green Vegetables Consumption and Fried Potato Consumption considered as continuous value since they range from 1 to 100 (some of them are up to 130).

Number of rows is: 308854

## Preliminary Descriptive Analysis of the Data

```python
# Representing statistics of numerical data:
numeric_cols = df[['Weight_(kg)','Alcohol_Consumption', 
         'Green_Vegetables_Consumption', 'FriedPotato_Consumption']]
numeric_cols.describe()
```

## Distributions of numerical columns: 

```python
# Plotting the histogram
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Weight_(kg)'], bins=30, kde=True)

# Adding labels and title
plt.title(f'Weight Distribution')
plt.xlabel('Weight [kg]')
plt.ylabel('Number of People')


plt.subplot(1, 2, 2)
sns.boxplot(x=df['Weight_(kg)'])
plt.title('Box Plot of Weight_kg')
plt.show()
```

### In the histogram below axis x is number of cups and axis y is number of people that consume the corresponding amount:

```python
# Plotting the histogram
plt.figure(figsize=(10, 5))
sns.histplot(df['Alcohol_Consumption'], bins=20, kde=True)

# Adding labels and title

plt.title(f'Alcohol Consumption Distribution')
plt.xlabel('Alcohol Consumption (Cups per month)')
plt.ylabel('Counts')


plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Alcohol_Consumption'])
plt.title('Box Plot of Alcohol_Consumption')
plt.show()
```

```python
# Plotting the histogram
plt.figure(figsize=(12, 4))
sns.histplot(df['Green_Vegetables_Consumption'], bins=45, kde=True)
#
# Adding labels and title
plt.title(f'Green Vegetables Consumption Distribution')
plt.xlabel('Green Vegetables Consumption per month')
plt.ylabel('Number of People')
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Green_Vegetables_Consumption'])
plt.title('Box Plot of Green Vegetables Consumption')
plt.show()
```

```python
# Plotting the histogram
plt.figure(figsize=(12, 5))
sns.histplot(df['FriedPotato_Consumption'], bins=25, kde=True)

# Adding labels and title
plt.title(f'Fried Potato Consumption Distribution')
plt.xlabel('Fried Potato Consumption per month')
plt.ylabel('Number of People')
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x=df['FriedPotato_Consumption'])
plt.title('Box Plot of FriedPotato_Consumption')
plt.show()
```

#### As We can see from the box plots, we have some outliers in the Weight category, Alcohol_Consumption, Green_Vegetables_Consumption and FriedPotato_Consumption.
####   But we can see from the histogram that they are a small proportion of the whole data.


### Represnting Categorical Variables 

```python
plt.figure(figsize=(10, 5))

# Bar plot for 'Depression'
ax1 = plt.subplot(1, 2, 1)
sns.countplot(x='Depression', data=df, palette='Set1')
plt.title('Bar Plot of Depression')
ax1.set_xticklabels(['No', 'Yes'])
plt.ylabel('Number of People')

# Bar plot for 'Sex'
ax2 = plt.subplot(1, 2, 2)
sns.countplot(x='Sex', data=df, palette='Set2')
plt.title('Bar Plot of Sex')
plt.ylabel('Number of People')
ax2.set_xticklabels(['Female', 'Male'])
plt.subplots_adjust(wspace=0.5)
plt.show()
```

```python
# Checking for missing data
missing_data = df.isnull()

# Counting the number of missing values in each column
missing_count = missing_data.sum()

# Displaying the columns with missing values and their counts
print("Columns with missing values:")
print(missing_count)
```

##### As we can see, there are no missing data.
